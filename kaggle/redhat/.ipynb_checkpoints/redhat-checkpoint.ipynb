```python
people_tmp = path + "people.csv"
people = pd.read_csv(people_tmp)

# train
act_train_tmp = path + "act_train.csv"
act_train = pd.read_csv(act_train_tmp)

# sample_sub
sample_tmp =  path + "sample_submission.csv"
sample_submission = pd.read_csv(sample_tmp)
```

```python
act_test_tmp = path + "act_test.csv"
act_test = pd.read_csv(act_test_tmp)
```

```python
act_train.head()
```

```python
activity_category = act_train['activity_category'].unique() # activity_categoryごとのリスト
```

```python
activity_category
```

```python
# 訓練データに切り分ける
train_feature = act_train.ix[:, 3:-2]
train_label = act_train.ix[:,-1]
```

```python
train_feature.head()
```

```python
# model 
model = RandomForestClassifier()
```

```python
# dumy変数化
dummy_actcate =  pd.get_dummies(train_feature["activity_category"]) 
# dummy_char10 =  pd.get_dummies(train_feature["char_10"]) 
```

```python
# concat
# drop col
train_feature = train_feature.drop("activity_category", axis=1)
```

```python
df = pd.concat((train_feature, dummy_actcate), axis=1)
```

```python
df.head()
```

```python
# dummy 変数化
# for 文で回す 
for col in train_feature.columns:
    dummy =  pd.get_dummies(train_feature[col]) 
    train_feature = pd.concat((train_feature, dummy), axis=1)
    train_feature =     train_feature.drop(col, axis=1)
```

```python
train_feature.head()
```

```python
model.fit(train_feature, train_label)
```

```python
act_test
```

```python
# 同様にtestデータに切り分ける
test_feature = act_test.ix[:, 3:-2]
test_label = act_test.ix[:,-1]
```

```python
for col in test_feature.columns:
    dummy =  pd.get_dummies(test_feature[col]) 
    test_feature= pd.concat((test_feature, dummy), axis=1)
    test_feature = test_feature.drop(col, axis=1)
```
