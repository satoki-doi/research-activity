ヒストグラムは、各カラーの明るさのレベル別にピクセル数をグラフ化し、画像内のピクセル分布を示したものです。
ヒストグラムには、シャドウ（ヒストグラムの左側部分）、中間調（ヒストグラムの中央）およびハイライト
（ヒストグラムの右側部分）のディテールが表示され、画像に適切な補正を行うための十分なディテールがあるかどうかを判断できます。

ヒストグラムからは、画像の色調範囲または画像のキータイプをすばやく識別することができます。
ローキー画像は、ディテールがシャドウに集中し、ハイキー画像はハイライトに集中します。アベレージキーの画像は、
中間調にディテールが集中します。色調範囲が全体に広がった画像は、全領域にピクセルが分布しています。
色調範囲を確認することは、適切な色調補正の判断に役立ちます。

```python
# 画像サイズ変更の関数
def imresize(im, sz):
    pil_im = Image.fromarray(uint8(im)) # 一度、画像形式に戻す必要あるみたいだ
    return array(pil_im.resize(sz))
```

```python
sample = imresize(test2_im, (128, 128))
```

```python
# PCAの実行
# flatten() 1次元ベクトルに変換する
from numpy import *

def pca(X):
    """
    主成分分析
    入力：X、訓練データを平板化した配列を業として格納した行列
    出力：写像行列（次元の重要度順）、分散、平均
    """
    # 次元数を取得する
    num_data, dim = X.shape
    
    # データをセンタリング
    mean_X = X.mean(axis=0)
    X = X - mean_X
    
    if dim > num_data:
        # PCA - 高次元の時はコンパクトな裏技を用いる
        M = dot(X, X.T) # 共分散行列
        e, EV = linalg.eigh(M) #固有値と固有ベクトル
        tmp = dot(X.T, EV).T # 裏技
        V = tmp[::-1]
        S = sqrt(e)[::-1]
        for i in range(V.shape[1]):
            V[:, i] /= S
    else:
        # PCA - 低次元なら特異値分解を用いる
        U, S, V = linalg.svd(X)
        V = V[:num_data] # 初めのnum_data分だけ有効

    return V, S, mean_X    
```

```python
# 使用
imlist = ['./test_data/4.jpg','./test_data/5.jpg','./test_data/6.jpg']
# im = array(Image.open('./test_data/4.jpg'))

#平板化
im_flatten_mat = array([imresize(array(Image.open(im)), (128, 128)).flatten()
                        for im in imlist], 'f')
```

```python
# 主成分分析の実行
V, S, im_mean = pca(im_flatten_mat)
```
